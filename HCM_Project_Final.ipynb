```python
import pandas as pd
import numpy as np
import geopy.distance
```

```python
df=pd.read_csv("AC Pop.csv")
```

```python
df
```

```python
# importing geopy library
from geopy.geocoders import Nominatim
 
# calling the Nominatim tool
loc = Nominatim(user_agent="GetLoc")
 
# entering the location name
getLoc = loc.geocode("Gosainganj Lucknow")
```

```python
latitude=[]
longitude=[]
address=[]

for i in df.iloc[:,1]:
  l=loc.geocode(i+' delhi')
  if l is not None:
    lat=l.latitude
    lon=l.longitude
    add=l.address

  else:
    lat=None
    long=None
    add=None

  latitude.append(lat)
  longitude.append(lon)
  address.append(add)
```

```python
df['Latitude']=latitude
df['Longitude']=longitude
df['Address']=address
```

```python
df
```

```python
df[df['Address'].isnull()]
```

```python
dis_mat=pd.DataFrame(index=df['AC Name'],columns=df['AC Name'])
```

```python
for i in range(len(df)):
  for j in range(len(df)):
    c1=(df.iloc[i,3],df.iloc[i,4])
    c2=(df.iloc[j,3],df.iloc[j,4])

    dis_mat.iloc[i,j]=np.round(float(geopy.distance.geodesic(c1, c2).km),decimals=2)
```
